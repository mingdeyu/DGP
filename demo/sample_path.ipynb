```python
from dgpsi import dgp, kernel, combine, lgp, path
import numpy as np
import matplotlib.pyplot as plt
```

Sample paths for 1-D input from a 3-layered DGP

```python
layer1=[kernel(length=np.array([1]),name='sexp')]
layer2=[kernel(length=np.array([1]),name='sexp')]
layer3=[kernel(length=np.array([1]),name='sexp')]
layer_all=combine(layer1,layer2,layer3)
```

```python
n=200
N=15
data=np.linspace(-1,1,n)[:,None]
syn=path(data,layer_all)
samp=syn.generate(N)
fig=plt.plot(data,samp[0,:].T)
```

Sample paths for 2-D input from a 3-layered DGP

```python
layer1=[kernel(length=np.array([1,2]),name='sexp')]
layer2=[kernel(length=np.array([1]),name='sexp')]
layer3=[kernel(length=np.array([1]),name='sexp')]
layer_all=combine(layer1,layer2,layer3)
```

```python
X = np.arange(0, 1, 0.02)
Y = np.arange(0, 1, 0.02)
X, Y = np.meshgrid(X, Y)
data=np.vstack((X.flatten(),Y.flatten())).T
syn=path(data,layer_all)
samp=syn.generate(1)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, np.reshape(samp[0,:].flatten(),np.shape(X)), cmap='viridis', edgecolor='none',
                       linewidth=0, antialiased=False,alpha=0.6)
ax.view_init(elev=40., azim=-60)
```
